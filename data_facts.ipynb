```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("hanoiweather.csv")
```

```python
# Basic information about the dataset
df.info()

# Descriptive statistics for the columns used in the project
columns_used = ['temp', 'humidity', 'windspeed', 'solarradiation', 'precip']
df[columns_used].describe()

#  Check for missing values in the columns used
missing_values = df[columns_used].isnull().sum()
print(missing_values)

# Create a binary rain column (similar to the project)
df['rain_binary'] = (df['precip'].fillna(0) > 0).astype(int)

# Value counts for the binary rain target
print("\nRain Distribution (Binary):")
rain_counts = df['rain_binary'].value_counts()
print(rain_counts)
print(f"Percentage of rainy days: {rain_counts[1] / len(df) * 100:.2f}%")

# Visualize the distribution of each feature
plt.figure(figsize=(15, 10))
for i, col in enumerate(columns_used):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

```
